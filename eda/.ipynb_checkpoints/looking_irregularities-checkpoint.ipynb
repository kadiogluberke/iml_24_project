```python
import mediapipe as mp
import numpy as np
import pandas as pd
import cv2
from matplotlib import pyplot as plt
```

```python
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=True, max_num_hands=2, min_detection_confidence=0.5, min_tracking_confidence=0.5)
mp_drawing = mp.solutions.drawing_utils
```

**None examples from none_list.txt**

```python
img_path = "../data/American Sign Language Letters.v6-raw.coco/train/O16_jpg.rf.8e148482e649dd4f55649ada7ec57227.jpg"
img = cv2.imread(img_path)
    
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img.flags.writeable = False

plt.imshow(img)
plt.show()
```

**Examples with landmarks more than 21**

```python
img_path = "../data/American Sign Language Letters.v6-raw.coco/valid/Y17_jpg.rf.bd6e5473ec7c16a36db08d3bddc66dd5.jpg"
img = cv2.imread(img_path)
    
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img.flags.writeable = False

plt.imshow(img)
plt.show()
```

```python
image = cv2.imread(img_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
results = hands.process(image_rgb)


if results.multi_hand_landmarks:
    for hand_landmarks in results.multi_hand_landmarks:
        mp_drawing.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
# plt.imshow(image)
plt.show()
```

```python
points = []
if results.multi_hand_landmarks:
    for hand_landmarks in results.multi_hand_landmarks:
        for i in range(21):
            x = hand_landmarks.landmark[i].x
            y = hand_landmarks.landmark[i].y
            z = hand_landmarks.landmark[i].z
            points.append([x, y, z])
```

```python
width = 400
height = 400

img = np.zeros([height, width, 3], dtype=np.uint8)
img.fill(255)
for hc in mp_hands.HAND_CONNECTIONS:
    cv2.line(img, (int((points[hc[0]][0]) * width),
                   int((points[hc[0]][1]) * height)),
                  (int((points[hc[1]][0]) * width),
                   int((points[hc[1]][1]) * height)), (0, 0, 255), 4)

plt.imshow(img)
plt.show()
```

```python
points = np.array(points)
```

```python
min_x = np.min(points[:, 0])
max_x = np.max(points[:, 0])
min_y = np.min(points[:, 1])
max_y = np.max(points[:, 1])
for point in points:
    point[0] = (point[0] - min_x) / (max_x - min_x)
    point[1] = (point[1] - min_y) / (max_y - min_y)
```

```python
width = 400
height = 400

img = np.zeros([height, width, 3], dtype=np.uint8)
img.fill(255)
for hc in mp_hands.HAND_CONNECTIONS:
    cv2.line(img, (int((points[hc[0]][0]) * width),
                   int((points[hc[0]][1]) * height)),
                  (int((points[hc[1]][0]) * width),
                   int((points[hc[1]][1]) * height)), (0, 0, 255), 4)

plt.imshow(img)
plt.show()
```
