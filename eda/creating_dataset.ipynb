```python
import pickle
import numpy as np
import pandas as pd
```

```python
with open('../data/datasets/data.pkl', "rb") as file:
    data = pickle.load(file)
```

```python
flattened_data = []
for instance in data:
    points = instance[:-1]
    label = instance[-1]
    flattened_instance = [item for sublist in points for item in sublist]
    flattened_instance.append(label)
    flattened_data.append(flattened_instance)
```

```python
# DataFrame column names
num_points = len(data[0]) - 1 
num_features = len(data[0][0])
columns = [f'landmark_{i}-axis_{j}' for i in range(num_points) for j in range(num_features)]
columns.append('label')

# DataFrame oluşturma
df = pd.DataFrame(flattened_data, columns=columns)

df.head()
```

```python
df.to_csv('../data/datasets/data.csv', index=False)
```
