```python
import pickle

import pandas as pd
import numpy as np

from sklearn import metrics, model_selection, ensemble

from tqdm import tqdm
```

```python
df_validation = pd.read_csv("../data/datasets/validation_data.csv")
df_validation.head()
```

```python
X_cols = [col for col in df_validation.columns if col != 'label']
```

```python
X_validation = df_validation[X_cols]
y_validation = df_validation.label
```

```python
with open('../models/random_forest_best/random_forest_best.pkl', 'rb') as file:
    clf = pickle.load(file)
```

```python
validation_predictions = clf.predict(X_validation)
```

```python
val_acc = metrics.accuracy_score(y_validation, validation_predictions)
```
